```python
from utils import get_openai_api_key
OPENAI_API_KEY = get_openai_api_key()
#print(OPENAI_API_KEY)
llm_config = {"model": "gpt-5-mini","api_key":OPENAI_API_KEY}
```

```python
from autogen import UserProxyAgent
```

```python
Planner= ConversableAgent( name ="Planner_Agent",
                           system_message= """You are a strategic business planner. When given a business idea, you:
                                                1. Create a comprehensive business plan outline.
                                                2. Identify specific questions for Market, Technical, and Financial experts
                                                3. Break down the plan into clear sections that each specialist should address. 
                                            Your response should not exceed 200 words
    """,
                           llm_config=llm_config,
                           human_input_mode="NEVER"
 )

Market = ConversableAgent( name= "Marketing_Agent",
                           system_message = """ You role is to provide the market analysis for the plan provided by the planner,
                                                 Your response should not exceed 100 words""",
                           llm_config=llm_config,
                           human_input_mode="NEVER"
                         )

Technical= ConversableAgent( name= "Technical_Agent",
                             system_message = """You are an technical expert, your role is to provide analysis for required technical
                                                  elements and infrastructure for the plan provide by the planner,
                                                  Your response should not exceed 100 words""",
                             llm_config=llm_config,
                             human_input_mode="NEVER"
                           )

Finance= ConversableAgent( name= "Finance_Agent",
                           system_message = """ You are a finance agent, you will provide financial analysis for the plan provided by the planner,
                                                Your response should not exceed 100 words""",
                           llm_config=llm_config,
                           human_input_mode="NEVER"
                         )

User_proxy= UserProxyAgent( name ="UserProxyAgent",
                            human_input_mode="ALWAYS",
                            code_execution_config=False)
                            
```

```python
consolidator = ConversableAgent(
    name="Business_Plan_Writer",
    system_message="""You are an expert business plan writer
     Make it cohesive, professional, and actionable.""",
    llm_config=llm_config,
    human_input_mode="NEVER"
)
```

```python
def reflection_message(recipient, messages, sender, config):
    history = recipient.chat_messages_for_summary(sender)
    content = history[-1]["content"]
    return f"Review the following content:\n\n{content}"
```

```python
chats = [
    {
        "sender": Planner,
        "recipient": Market,
        "message": reflection_message,
        #"summary_method": "reflection_with_llm",
        #"summary_args": {
            #"summary_prompt": "Return the customer information as JSON object only: {'name': '', 'location': ''}",
        #},
        "max_turns": 1,
        #"clear_history": True
    },
    {
        "sender": Planner,
        "recipient": Technical,
        "message": reflection_message,
        #"summary_method": "reflection_with_llm",
        "max_turns": 1,
        #"clear_history": False
    },
    {
        "sender": Planner,
        "recipient": Finance,
        "message": reflection_message,
        #"summary_method": "reflection_with_llm",
        "max_turns": 1,
        #"clear_history": False
    },
        {
        "sender": Planner,
        "recipient": consolidator,
        "message": reflection_message,
        "carryover":[0,1,2],
        "summary_method": "reflection_with_llm",
        "max_turns": 1,
        #"clear_history": False
    },
]
```
