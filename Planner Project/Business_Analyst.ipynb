```python
Planner= ConversableAgent( name ="Planner_Agent",
                           system_message= """You are a strategic business planner.
    
    When given a business idea, you:
    1. Create a comprehensive business plan outline
    2. Identify specific questions for Market, Technical, and Financial experts
    3. Break down the plan into clear sections that each specialist should address. Keep it in 100 words""",
                           llm_config=llm_config,
                           human_input_mode="NEVER"
 )

Market = ConversableAgent( name= "Marketing_Agent",
                           system_message = " You role is to provide the market analysis for the plan provided by the planner, Keep it in 100 words",
                           llm_config=llm_config,
                           human_input_mode="NEVER"
                         )

Technical= ConversableAgent( name= "Technical_Agent",
                             system_message = """You are an technical expert, your role is to provide analysis for required technical
                                               elements and infrastructure for the plan provide by the planner, Keep it in 100 words""",
                             llm_config=llm_config,
                             human_input_mode="NEVER"
                           )

Finance= ConversableAgent( name= "Finance_Agent",
                           system_message = """ You are a finance agent, you will provide financial analysis for the plan provided by the planner,
                           Keep it in 100 words""",
                           llm_config=llm_config,
                           human_input_mode="NEVER"
                         )

User_proxy= UserProxyAgent( name ="UserProxyAgent",
                            human_input_mode="ALWAYS",
                            code_execution_config=False)
                            
```

```python
def get_plan_for_reflection(recipient, messages, sender, config):
    """Custom message that references the actual plan"""
    return f"""Review this business plan and provide your expert analysis:{Plan}
    Focus on your area of expertise and provide specific, actionable recommendations."""



nested_chats = [
    {
        "sender": Planner,
        "recipient": Market,
        "message": get_plan_for_reflection,
        "summary_method": "reflection_with_llm",
        "max_turns": 1,
    },
    {
        "sender": Planner,
        "recipient": Technical,
        "message": get_plan_for_reflection,
        "summary_method": "reflection_with_llm",
        "max_turns": 1,
    },
    {
        "sender": Planner,
        "recipient": Finance,
        "message": get_plan_for_reflection,
        "summary_method": "reflection_with_llm",
        "max_turns": 1,
    },
]

# Now register the nested chats
Planner.register_nested_chats(nested_chats, trigger=User_proxy)

# Trigger the nested consultations with a follow-up message
followup_chat = Planner.initiate_chat(
    recipient=User_proxy,
    message="Now let me consult our specialists to refine this plan...",
    max_turns=2,
)
```
