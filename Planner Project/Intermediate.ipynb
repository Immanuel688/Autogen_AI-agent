```python
import autogen
from autogen import ConversableAgent
from utils import get_openai_api_key
OPENAI_API_KEY = get_openai_api_key()
#print(OPENAI_API_KEY)
llm_config = {"model": "gpt-5-mini","api_key":OPENAI_API_KEY}
from autogen import UserProxyAgent
```

```python
Planner= ConversableAgent( name ="Planner_Agent",
                           system_message= """You are a strategic business Planner.
    
    When given a business idea, you:
    1. Create a comprehensive business plan outline
    2. Identify specific questions for Market, Technical, and Financial experts
    3. Break down the plan into clear sections that each specialist should address. Keep it in 100 words""",
                           llm_config=llm_config,
                           human_input_mode="NEVER"
 )

Market = ConversableAgent( name= "Marketing_Agent",
                           system_message = " You role is to provide the market analysis for the plan provided by the Planner, Keep it in 100 words",
                           llm_config=llm_config,
                           human_input_mode="NEVER"
                         )

Technical= ConversableAgent( name= "Technical_Agent",
                             system_message = """You are an technical expert, your role is to provide analysis for required technical
                                               elements and infrastructure for the plan provide by the Planner, Keep it in 100 words""",
                             llm_config=llm_config,
                             human_input_mode="NEVER"
                           )

Finance= ConversableAgent( name= "Finance_Agent",
                           system_message = """ You are a finance agent, you will provide financial analysis for the plan provided by the Planner,
                           Keep it in 100 words""",
                           llm_config=llm_config,
                           human_input_mode="NEVER"
                         )

User_proxy= UserProxyAgent( name ="UserProxyAgent",
                            human_input_mode="ALWAYS",
                            code_execution_config=False)

consolidator = ConversableAgent(
    name="Business_Plan_Writer",
    system_message="""You are an expert business plan writer
    
    Make it cohesive, professional, and actionable.""",
    llm_config=llm_config,
    human_input_mode="NEVER"
)
```

```python
# Planner talks to user to get the idea
result = User_proxy.initiate_chat(
    recipient=Planner,
    #message="What's your business idea?",
    max_turns=1
)
```

```python
# Planner talks to Tech agent
tech_result = Planner.initiate_chat(
    recipient=Technical,
    message=f"Technical plan for: {result.summary}",
    max_turns=1
)
```

```python

# Planner talks to Finance agent
finance_result = Planner.initiate_chat(
    recipient=Finance,
    message=f"Financial projection for: {result.summary}",
    max_turns=1
)
```

```python

# Consolidator talks to user
final_result = consolidator.initiate_chat(
    recipient=User_proxy,
    message=f"Final plan combining: {market_result.summary}, {tech_result.summary}, {finance_result.summary}",
    max_turns=1
)
```

```python
# YOUR MENTAL MODEL:
# 1. User provides idea
# 2. Planner creates outline
# 3. Market, Tech, Finance analyze in parallel
# 4. Consolidator synthesizes

# MAPS TO:
# Sequential conversations with parallel analysis (using carryover)

# USE THIS METHOD:
from autogen import initiate_chats

workflow = [
    # Step 1: Get idea and create plan
    {
        "sender": User_proxy,
        "recipient": Planner,
        "message": "I want to create a mental health startup",
        "max_turns": 1,
        "summary_method": "last_msg"
    },
    
    # Steps 2-4: Parallel specialist analyses
    {
        "sender": Planner,
        "recipient": Market,
        "message": "Market analysis",
        "carryover": [0],  # Get the plan
        "max_turns": 1
    },
    {
        "sender": Planner,
        "recipient": Technical,
        "message": "Technical requirements",
        "carryover": [0],
        "max_turns": 1
    },
    {
        "sender": Planner,
        "recipient": Finance,
        "message": "Financial projections",
        "carryover": [0],
        "max_turns": 1
    },
    
    # Step 5: Consolidate
    {
        "sender": User_proxy,
        "recipient": consolidator,
        "message": "Create final business plan",
        "carryover": [0, 1, 2, 3],  # Get everything
        "max_turns": 1
    }
]

results = initiate_chats(workflow)
```
